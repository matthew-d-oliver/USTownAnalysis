```python
def get_fax(town):
    
    enter_text=driver.find_element_by_xpath('//*[@id="cfsearchtextbox"]')
    enter_text.clear()
    enter_text.send_keys(town)
    driver.find_element_by_xpath('//*[@id="cfsearchform"]/a').click()
    wait=WebDriverWait(driver, delay)
    time.sleep(2.5)
    try:
        driver.find_element_by_xpath('//*[@id="_modal_container_"]/div[1]/div/div/alert-modal/div[3]/button').click()
        print('no result')
        return np.nan
    except:
        try:
            
#             page = wait.until(EC.presence_of_element_located((By.XPATH,'//*[@id="cf-content"]/div[1]/div/table/tbody/tr[1]/td')))
            try:
                temp = driver.find_element_by_css_selector('tbody').text.split('Median Household Income')[1]
                temp = temp.lstrip(' ')
                income=temp.split(' ', 1)[0]
            except:
                income=np.nan
            try:
                temp = driver.find_element_by_css_selector('tbody').text.split('Educational Attainment: Percent high school graduate or higher')[1]
                temp = temp.lstrip(' ')
                learn=temp.split(' ', 1)[0]
            except:
                edu=np.nan
            try:
                temp = driver.find_element_by_css_selector('tbody').text.split('Median Age')[1]
                temp = temp.lstrip(' ')
                age=temp.split(' ', 1)[0]
            except:
                age=np.nan
            try:
                temp = driver.find_element_by_css_selector('tbody').text.split('Individuals below poverty level')[1]
                temp = temp.lstrip(' ')
                poverty=temp.split(' ', 1)[0]
            except:
                poverty=np.nan
            town_data={'name':town,'income':income, 'age':age, 'poverty': poverty, 'edu':learn}
            return town_data
        except TimeoutException:
            print('too long')
            return np.nan
    
        
```
